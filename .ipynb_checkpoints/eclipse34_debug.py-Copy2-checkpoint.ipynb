```python
from scipy.io import arff
import pandas as pd
import numpy as np
import math

from numpy import unique
from numpy import where
from matplotlib import pyplot

data = arff.loadarff('./DefectData/inst/extdata/kim/eclipse34_swt.arff') 
df = pd.DataFrame(data[0])
```

```python
df.head()
```

```python
x = df.drop(['isDefective'],axis = 'columns')
x.head()
```

```python
x.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()
scalar.fit(x)
x = scalar.transform(x)
```

```python
x = pd.DataFrame(x)
x.head()
```

```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters = 2)
kmeans_model.fit(x)

kmeans_prediction = kmeans_model.predict(x)
x['cluster'] = kmeans_prediction

kmeans_prediction[:5]
```

```python
def ASFM(DataFrame):
    SFM = 0
    
    x,y = DataFrame.shape
    
    for i in range(x):
        for j in range(y-1):
            SFM += DataFrame.iloc[i,j]
    
    return SFM/x
```

```python
df0 = x[x.cluster==0]
df1 = x[x.cluster==1]
```

```python
ASFM0 = ASFM(df0)
ASFM1 = ASFM(df1)

print(f"ASFM0 : {ASFM0}")
print(f"ASFM1 : {ASFM1}")
```

```python
#ASFM0 value is greater than ASFM1, so label 1 will represent defected software and 0 represents non-defected 
#software



TP = 0
FN = 0
TN = 0
FP = 0


for i in range(1485):
    
    defect = df.iloc[i,17]
    buggy = b'FALSE'
    clean = b'TRUE'
    
    if defect == buggy and kmeans_prediction[i]==0:
        TP += 1
    elif defect == buggy and kmeans_prediction[i]==1:
        FN += 1
    elif defect == clean and kmeans_prediction[i]==1:
        TN += 1
    else:
        FN += 1
 
```

```python
#Matthew Correlation Coefficient

MCC = (TP*TN-FP*FN)/math.sqrt((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN))

accuracy = (TP+TN)/len(df)

print(f"MCC : {MCC}")
print(f"Accuracy : {accuracy}")
```

```python
from sklearn.cluster import AgglomerativeClustering

ag_model = AgglomerativeClustering(n_clusters = 2)

x = x.drop(['cluster'],axis = 'columns')
ag_model.fit(x)
ag_prediction = ag_model.labels_
x['cluster'] = ag_prediction

ag_prediction[:5]
```
